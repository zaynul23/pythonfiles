```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2

mousd=np.ones((512,512,3))
# mousd*=255
# photoedit=cv2.imread(imgad)
# mousd=cv2.imread(imgad)

def drawcircle(butp,x,y,flags,param):
    if butp ==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(mousd,(x,y),25,(0,255,0),thickness=-1)

cv2.namedWindow("my drawing")
cv2.setMouseCallback("my drawing",drawcircle)
    
while True:
    cv2.imshow("my drawing",mousd)
    if cv2.waitKey(20)==ord("q"):
        break   
cv2.destroyAllWindows()
```

```python
import cv2,time


cap=cv2.VideoCapture(r"foler_w_vids\lapcam.mp4")

if cap.isOpened()==False:
    print("file not found")
start=time.time()
while cap.isOpened():
    
    ret,frame=cap.read()
    
    
    if ret:
        time.sleep(1)
        cv2.imshow("frame",frame)
        
        if cv2.waitKey(1)==ord("q"):
            break
    else:
        break
print(f"clip length is {time.time()-start}")
cv2.destroyAllWindows()
cap.release()
```

```python
import cv2

pt1=(0,0)
pt2=(0,0)
topLclked=False
bottomRclked=False

cap=cv2.VideoCapture(0)

def draw_rectangle(event,x,y,flags,param):
    global pt1,pt2,topLclked,bottomRclked
    if event == cv2.EVENT_LBUTTONUP:
        if topLclked and bottomRclked:
            pt1=pt2=(0,0)
            topLclked=bottomRclked=False
        

        if not topLclked:
            pt1=(x,y)
            topLclked=True
        elif not bottomRclked:
            pt2=(x,y)
            bottomRclked=True    

cv2.namedWindow("Test")
cv2.setMouseCallback("Test",draw_rectangle)


while True:
    
    ret,frame=cap.read()
    
    if topLclked:
        cv2.circle(frame,pt1,5,(0,255,0),-1)
        
    if topLclked and bottomRclked:
        cv2.rectangle(frame,pt1 ,pt2,(0,0,255),4)
    
    cv2.imshow("frame",frame)
    
    if cv2.waitKey(1) ==ord('q'):
        break
    
cv2.destroyAllWindows()
cap.release()
```

```python
methods=["cv2.TM_CCOEFF","cv2.TM_CCOEFF_NORMED","cv2.TM_CCORR","cv2.TM_CCORR_NORMED","cv2.TM_SQDIFF","cv2.TM_SQDIFF_NORMED"]
```

```python
import cv2
for m in methods:
    fullcopy=full.copy()
    halfcopy=half.copy()

    method=eval(m)
    
    res=cv2.matchTemplate(fullcopy,halfcopy,method)
    min_v,max_v,min_l,max_l=cv2.minMaxLoc(res)
    
    if method in ["cv2.TM_SQDIFF","cv2.TM_SQDIFF_NORMED"]:
        top_left=min_l
    else:
        top_left=max_l
    
    height,width,channels=half.shape
    
    bototmright=(top_left[0]+width,top_left[1]+height)
    
    cv2.rectangle(fullcopy,top_left,bototmright,(255,0,0),4)
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title("heatmap of fred")
    
    plt.subplot(122)
    plt.imshow(fullcopy)
    plt.title("detectof template")
    
    plt.subplot(121)
    plt.imshow(halfcopy)
    plt.title("half wala")
    
    plt.suptitle(m)
    
    plt.show()
    
    print('\n'*3)   
    
    
```

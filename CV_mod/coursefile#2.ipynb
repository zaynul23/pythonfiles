                                                FLAN MATCHING

```python
import cv2

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cf=r"D:\zaynu\Documents\pythonfiles\CV_mod\Computer-Vision-with-Python\DATA\reeses_puffs.png"
C_Ap=r"D:\zaynu\Documents\pythonfiles\CV_mod\Computer-Vision-with-Python\DATA\many_cereals.jpg"
cereal=cv2.imread(cf,0)
aislecereal=cv2.imread(C_Ap,0)
plt.imshow(aislecereal,cmap='gray')
```

```python
sift=cv2.SIFT_create()
kp1,des1=sift.detectAndCompute(cereal,None)
kp2,des2=sift.detectAndCompute(aislecereal,None)
```

```python
flan_index_kdtree=0
index_params= dict(algorithm=flan_index_kdtree,trees=5)
searchparams=dict(checks=50)
```

```python
flann=cv2.FlannBasedMatcher(index_params,searchparams)
```

```python
matche=flann.knnMatch(des1,des2,k=2)
```

```python
good=[]
for m1,m2 in matche:
    if m1.distance < 0.7*m2.distance:
        good.append([m1])
```

```python
flann_mactches=cv2.drawMatchesKnn(cereal,kp1,aislecereal,kp2,good,None,flags=2)
#flag 0 leads to circles ?? 
```

```python
# outres=cv2.cvtColor(flann_mactches,cv2.COLOR_RGB2BGR)
# cv2.imshow('flanmatching',outres)
# cv2.waitkey(0)
# cv2.destroyAllWindows()
outrres=cv2.cvtColor(flann_mactches,cv2.COLOR_RGB2BGR)
cv2.imshow("res",outrres)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
matchs=flann.knnMatch(des1,des2,k=2)
matchmask=[ [0,0] for i in range(len(matchs))]

for i ,(m1,m2) in enumerate(matchs):
    if m1.distance < m2.distance *0.7:
        matchmask[i]=[1,0]
```

```python
draw_params=dict(matchColor=(0,255,0),singlePointColor=(255,0,0),matchesMask=matchmask,flags=0)
```

```python
flann_mactches=cv2.drawMatchesKnn(cereal,kp1,aislecereal,kp2,matchs,None,**draw_params)
# cv2.imshow('outres',flann_mactches)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
outrres=cv2.cvtColor(flann_mactches,cv2.COLOR_RGB2BGR)
cv2.imshow("res",outrres)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#optional

kernal=np.ones((3,3),np.uint8)
opninig=cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernal,iterations=2)
sure_bg=cv2.dilate(opninig,kernal,iterations=3)
show(sure_bg)
```

```python
dist_transform=cv2.distanceTransform(opninig,cv2.DIST_L2,5)
show(dist_transform)
```

```python
ret,sure_fg=cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
show(sure_fg)
```

```python
# 
sure_fg=np.uint8(sure_fg)
unkown=cv2.subtract(sure_bg,sure_fg)
```

```python
ret,markers=cv2.connectedComponents(sure_fg)
show(markers)
markers=markers+1
show(markers)
markers[unkown==255]=0
show(markers)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import cm


def show(img,cma=['gray']):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,*cma)
```

```python
#simplified watershed

road=cv2.imread(r'Computer-Vision-with-Python\DATA\road_image.jpg')
roadcop=np.copy(road)
markerimg=np.zeros(shape=road.shape[:2],dtype=np.int32) 
segments=np.zeros(shape=road.shape,dtype=np.uint8)
markerimg_copy=markerimg.copy()
colormark=np.zeros(shape=road.shape[:2],dtype=np.uint8)

# show(markerimg_copy)
# plt.imshow(road)
 
```

```python
# cm.tab10(0)

def create_rgb(i):
    return tuple(np.array(cm.tab10(i)[:3])*255)

funccolors=[create_rgb(i) for i in range(10)]
funccolors
```

```python
current_marker=1
markupdate=False
def mouse_callback(event,x,y,flags,param):
    global markupdate
    
    if event==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(markerimg,(x,y),10,(current_marker),-1)
        
        cv2.circle(roadcop,(x,y),10,funccolors[current_marker],-1)
        markupdate=True
# cv2.imshow('markingimage',markerimg)
    
    
```

```python

cv2.namedWindow('ROAD Image')
cv2.setMouseCallback('ROAD Image',mouse_callback)
colormark=np.zeros(shape=markerimg.shape)

while True:
    cv2.imshow('Watershed Segements',segments)
    cv2.imshow('ROAD Image',roadcop)
    # cv2.imshow('markerimg',markerimg_copy)
    cv2.imshow('markingimage',colormark)
    k=cv2.waitKey(1)
    
    if k==ord('q'):
        break

    elif k==ord('c'):
        roadcop=road.copy()
        markerimg=np.zeros(shape=roadcop.shape[:2],dtype=np.int32)
        segments=np.zeros(shape=roadcop.shape,dtype=np.uint8)
        markupdate=False
    
    elif 47<k<58:
        current_marker=int(chr(k))
        
    if  markupdate:
        
        markerimg_copy=markerimg.copy()
        cv2.watershed(road,markerimg_copy)
        segments= np.zeros(road.shape,dtype=np.uint8)
        
        for colorid in range(len(funccolors)):
            
            segments[markerimg_copy==(colorid)]=funccolors[colorid]
            colormark[markerimg_copy==(colorid)]=colorid
            
        
cv2.destroyAllWindows()
     
```

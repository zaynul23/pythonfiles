```python
def repace_location():
    replacing_choice='erfqf'
    while replacing_choice not in range(6):
        replacing_choice=int(input('enetr your index to replace item (0-5) '))
        if replacing_choice not in range(6):
            print('sorry invalid input')
    return (replacing_choice)
```

```python
def replacement_choice(gamelist,position):
    userinput=input('enetr your item to replace at location ')
    gamelist[position]=userinput
    return gamelist
```

```python
def board(symbols):
    print('\n'*100)
    print('           |           |           \n'
        '           |           |           \n'
        f'     {symbols[1]}     |     {symbols[2]}     |     {symbols[3]}      \n'
        '           |           |           \n'
        '           |           |           \n'
        '-----------|-----------|-----------\n'
        '           |           |           \n'
        '           |           |           \n'             
        f'     {symbols[4]}     |     {symbols[5]}     |     {symbols[6]}      \n'
        '           |           |           \n'
        '           |           |           \n'
        '-----------|-----------|-----------\n'
        '           |           |           \n'
        '           |           |           \n'             
        f'     {symbols[7]}     |     {symbols[8]}     |     {symbols[9]}      \n'
        '           |           |           \n'
        '           |           |           ')
    

    
    
def player_selec():
    p1='lols'
    while p1 not in ['O','X','x','o']:
        p1=input('please enter what symbol you are (X/O): ')
    if p1.lower()=="o":
        return 'O'
    else:
        return 'X'

    
    
    
def game_on(status):
    if status:
        status_in_fucny='fef'
        while status_in_fucny not in['Y','N']:
            status_in_fucny=input('would you like to continue the game(Y/N)')
            if status_in_fucny not in['Y','N']:
                print('please enter valid answer (Y/N)')
        if status_in_fucny=='Y':
            status=True
            return status
        else:
            status=False
            return status

    
    
    
    
def clear_output():
    print('\n'*100)
    
    
    
    
def player_choice(symbols,p1,p2):
    global turn
    if turn%2!=0:
        status_in=True
        while status_in:
            choice='lingff'
            while choice not in [1,2,3,4,5,6,7,8,9]:
                choice=int(input('player 1 please enter position u want to enter ur piece'))
            if symbols[choice]!=' ':
                print('that spot is already filled try again')
            else:
                symbols[choice]=p1
                status_in=False
                turn+=1
                return symbols
    else:
        status_in=True
        while status_in:
            choice='lingff'
            while choice not in [1,2,3,4,5,6,7,8,9]:
                choice=int(input('player 2 please enter position u want to enter ur piece'))
            if symbols[choice]!=' ':
                print('that spot is already filled try again')
            else:
                symbols[choice]=p2
                status_in=False
                turn+=1
                return symbols

            
            
            
            
            
            
            
            
def wincheck(symbols,p1):
    x=1
    while x<8:
        if symbols[x]==symbols[x+1]==symbols[x+2]!=' ':
            if symbols[x]==p1:
                return 'player 1 has won'
            else:
                return 'player 2 has won'
        else:
            x+=3
    x=1
    while x<4:
        if symbols[x]==symbols[x+3]==symbols[x+6]!=' ':
            if symbols[x]==p1:
                return 'player 1 has won'
            else:
                return 'player 2 has won'
        else:
            x+=1
    if symbols[1]==symbols[5]==symbols[9]!=' ':
        if symbols[x]==p1:
            return 'player 1 has won'
        else:
            return 'player 2 has won'
    elif symbols[3]==symbols[5]==symbols[7]!=' ':
        if symbols[x]==p1:
            return 'player 1 has won'
        else:
            return 'player 2 has won'
    else:
        return None

        
                
    
```

```python


                                            #"the start"
symbols=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
status=True
turn=1
p1=player_selec()
p2='fa'
if p1=='O':
    p2='X'
else:
    p2='O'
while status:    
    board(symbols)
    symbols=player_choice(symbols,p1,p2)
    win_condition=wincheck(symbols,p1)
    board(symbols)
    if win_condition!=None:
        print(win_condition)
        status=game_on(status)
        if status:
            symbols=['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
            turn=1
            p1=player_selec()
            p2='fa'
            if p1=='O':
                p2='X'
            else:
                p2='O'
    if status:
        clear_output()

```
